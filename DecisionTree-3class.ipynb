```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
x = [[27],
     [35],
     [39],
     [21],
     [22],
     [45],
     [102],
     [110],
     [120]]

y_name = ["lemon", "orange", "orange", "lemon", "lemon", "orange", "pamelo", "pamelo", "pamelo"]
y = [0, 1, 1, 0, 0, 1, 2, 2, 2]
```

```python
plt.scatter(y_name, x)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x, y)
```

```python
x_input = [[25.3]]
y_predict = model.predict(x_input)
print(y_predict)
```

```python
fruit_class = ["lemon", "orange", "pamelo"]

prd = fruit_class[y_predict[0]]
print("Result : {}".format(prd))
```

```python
x_input = [[31.0],
           [25.0],
           [21.2],
           [38.0],
           [101]]

y_predict = model.predict(x_input)
print(y_predict)
```

```python
for i, val in enumerate(y_predict):
    print(f"{x_input[i]} = {val} {fruit_class[val]}")
```

```python
from sklearn.tree import plot_tree

data_feature_name = ['size']

plt.subplots(nrows=1 ,ncols=1, figsize=(5,4), dpi=96)

plot_tree(model,
          filled=True, rounded=True,
          feature_names=data_feature_name,
          class_names=fruit_class,)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
x = np.array(x)
y = np.array(y)

plot_decision_regions(x, y, clf=model)
```
